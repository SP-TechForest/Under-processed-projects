```python
needed_col = list(set(corr_matrix["SalePrice"][(corr_matrix["SalePrice"] > 0.50) | (corr_matrix["SalePrice"] < -0.50)].index) & set(corr_matrix["SalePrice"][(corr_matrix["SalePrice"] < 0.85) | (corr_matrix["SalePrice"] > -0.85)].index))
```

```python
cat_cols = ["MSZoning", "Utilities","BldgType","Heating","KitchenQual","SaleCondition","LandSlope"]
```

```python
needed_col
```

```python
neede_catdata = data[cat_cols]
```

```python
neede_catdata = pd.get_dummies(neede_catdata)
```

```python
final_data = final_data.join(neede_catdata)
```

```python
final_data
```

```python
final_data.columns
```

```python
final_data.columns[final_data.isna().any()]
```

```python
feature_data = final_data.drop("SalePrice", axis = 1)
label_data = final_data["SalePrice"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature_data, label_data, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
tf.random.set_seed(20)

prediction_model3 = tf.keras.Sequential([
                                        tf.keras.layers.Dense(100),
                                        tf.keras.layers.Dense(10),
                                        tf.keras.layers.Dense(1)
])

prediction_model3.compile(
    loss = tf.keras.losses.mape,
    optimizer = tf.keras.optimizers.SGD(),
    metrics = ["mape"] 
)

prediction_model3.fit(x_train, y_train, epochs=100)
```

```python
y_pred_model1 = prediction_model3.predict(x_test)
x_test
```

```python
prediction_model3.summary()
```

```python
loss_val = tf.keras.metrics.mean_absolute_error(y_test, y_pred_model1.squeeze())
loss_val.numpy()
```

```python
result = pd.DataFrame()
result["Actual"] = y_test
result["Pridected"] = y_pred_model1
```
